```python
#Import necessary Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data=pd.read_csv("/content/data.csv",encoding='latin-1')
data30=pd.read_csv("/content/data.csv",encoding='latin-1',nrows=30)
data100=pd.read_csv("/content/data.csv",encoding='latin-1',nrows=100)
data500=pd.read_csv("/content/data.csv",encoding='latin-1',nrows=500)
data.head()
```

```python
data.shape
```

```python
wagesort=data30['Wage'].sort_values()
plt.figure(figsize=(15,15))
plt.hist(wagesort)
plt.xlabel("wage in k")
plt.title("Wages of Players")
plt.show()
```

```python
age=data['Age']
potential=data['Potential']

plt.bar(age, potential)
plt.title("potential of players based on their age") 
plt.xlabel("Age") 
plt.ylabel("potential on 100")
plt.show()
```

```python
data1=data.head()
age=data1['Age']
acc=data1['Acceleration']
speed=data1['SprintSpeed']
agility=data1['Agility']

plt.style.use('seaborn-bright')
plt.plot(age, acc, label = "Accelaration")
plt.plot(age, speed, label = "sprint speed")
plt.plot(age, agility, label = "agility")
plt.title("Age vs capabilities")
plt.xlabel("Age")
plt.ylabel("capability")
plt.legend()
plt.show()
```

```python
data30=pd.read_csv("/content/data.csv",encoding='latin-1',nrows=30)
name=data30['Name']
strength=data30['Strength']
plt.figure(figsize=(8,6))
plt.barh(name, strength)
plt.xlabel("strength")
plt.title("Strength of top 30 players")
plt.show()
```

```python
data['Body Type'].value_counts()
```

```python
plt.title("Body-Type of players")
slices=[10598,6419,1142]
labels=['Normal','Lean','Stocky']
plt.pie(slices,labels=labels,shadow=True,autopct="%0.1f%%",wedgeprops={'edgecolor':'black'},radius=1)
plt.show()
```

```python
height=data30['Height']
weight=data30['Weight']
plt.figure(figsize=(8,6))
plt.scatter(height, weight,cmap='summer', edgecolors='black', linewidths=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label("BMI")
plt.xlabel("height")
plt.ylabel("weight")
plt.title("Body Mass Index")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(data['Preferred Foot'], )
plt.title('Most Preferred Foot of the Players', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (18, 8))
plt.style.use('fivethirtyeight')
ax = sns.countplot('Position', data = data, palette = 'inferno')
ax.set_xlabel(xlabel = 'Different Positions in Football', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Players', fontsize = 16)
ax.set_title(label = 'Comparison of Positions and Players', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
ax = sns.countplot(x = 'Skill Moves', data = data, palette = 'coolwarm')
ax.set_title(label = 'Count of players on Basis of their skill moves', fontsize = 20)
ax.set_xlabel(xlabel = 'Number of Skill Moves', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (13, 8))
ax = sns.countplot(x = 'Height', data = data, palette = 'gist_rainbow')
ax.set_title(label = 'Count of players on Basis of Height', fontsize = 20)
ax.set_xlabel(xlabel = 'Height in Foot per inch', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```
